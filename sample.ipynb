Libraries

```python
import pandas as pd
from pycoingecko import CoinGeckoAPI
import plotly.graph_objects as go
import plotly
```

Begin code

```python
cg = CoinGeckoAPI()
bitcoin=cg.get_coin_market_chart_by_id(id="bitcoin",vs_currency="usd",days=30)

```

```python
data = pd.DataFrame(bitcoin['prices'],columns=['timestamp','price'])
```

```python
data['Date'] = pd.to_datetime(data['timestamp'], unit='ms')
```

```python
candle = data.groupby(data.Date.dt.date).agg({'price':['min','max','first','last']})
```

```python
fig = go.Figure(data=go.Candlestick(x=candle.index,
                                    open=candle['price']['first'],
                                    high=candle['price']['max'],
                                    low=candle['price']['min'],
                                    close=candle['price']['last']))
fig.update_layout( xaxis_rangeslider_visible=True,xaxis_title='Date',yaxis_title='Price(USD)',title='Candlestick Price')
```
